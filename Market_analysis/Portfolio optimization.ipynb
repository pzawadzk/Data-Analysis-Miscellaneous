```python
from math import sqrt
from cvxopt import matrix
from cvxopt.blas import dot
from cvxopt.solvers import qp
import datetime
import pandas as pd
import pandas.io.data
```

```python
df = pd.io.data.get_data_yahoo(['FUSEX', 'IVE', 'IJS', 'FSIIX'], 
                               start=datetime.datetime(2000, 8, 1), 
                               end=datetime.datetime(2014, 11, 15))['Adj Close']
df.head()
```

```python
ret = df.pct_change()
print 'STD\n', ret.std()
pbar = matrix([df[col][-1]/df[col][0]-1 for col in df.columns])
print 'Return\n', pbar
```

```python
S = matrix(ret.cov().values)
print S
```

```python

# Problem data.
n = df.shape[1]

G = matrix(0.0, (n,n))
G[::n+1] = -1.0
h = matrix(0.0, (n,1))
A = matrix(1.0, (1,n))
b = matrix(1.0)

# Compute trade-off.
N = 100
mus = [ 10**(5.0*t/N-1.0) for t in range(N) ]
portfolios = [ qp(mu*S, -pbar, G, h, A, b)['x'] for mu in mus ];
returns = [ dot(pbar,x) for x in portfolios ]
risks = [ sqrt(dot(x, S*x)) for x in portfolios ]
```
