```python
import datetime

import pandas as pd
import pandas.io.data
%pylab inline


```

```python
df = pd.io.data.get_data_yahoo(['AAPL', 'GE', 'IBM', 'KO', 'MSFT', 'PEP', '^GSPC'], 
                               start=datetime.datetime(2010, 1, 1), 
                               end=datetime.datetime(2014, 10, 20))['Adj Close']
df.head()
```

```python
close_px = df['^GSPC']
```

```python
ret = df.pct_change()
```

```python
plot(ret['^GSPC'], ret['AAPL'], '*')
ax = gca()
ax.set_xlabel('GSPC')
ax.set_ylabel('APPL')
```

```python
pd.scatter_matrix(ret, diagonal='kde', figsize=(10, 10));
```

```python
corr = ret.corr()
corr
```

```python
plt.imshow(corr, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns);
```

```python
ax = gca()
ax.scatter(ret.mean(), ret.std())
ax.set_xlabel('Expected returns')
ax.set_ylabel('Risk')
for label, x, y in zip(ret.columns, ret.mean(), ret.std()):
    ax.annotate(
        label, 
        xy = (x, y), xytext = (20, -20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
ax.axis(xmin=-0.001, xmax=0.002, ymin=0.005, ymax=0.02)
```
