```python
from math import sqrt
from cvxopt import matrix
from cvxopt.blas import dot
from cvxopt.solvers import qp
import datetime
import pandas as pd
import pandas.io.data

%pylab inline
```

```python
df = pd.io.data.get_data_yahoo(['AAPL', 'GE', 'TEDIX', 'SPLV'], 
                               start=datetime.datetime(2011, 10, 20), 
                               end=datetime.datetime(2014, 10, 20))['Adj Close']
df.head()
```

```python
ret = df.pct_change()
print 'STD\n', ret.std()
pbar = matrix([df[col][-1]/df[col][0]-1 for col in df.columns])
print 'Return\n', pbar
```

```python
S = matrix(ret.cov().values)
print S
```

```python
from IPython.core.display import Image 
Image(url='http://cvxopt.org/userguide/_images/math/305efdce8b67069139cfdce108379dd0f9c13e14.png') 
```

```python

# Problem data.
n = df.shape[1]

G = matrix(0.0, (n,n))
G[::n+1] = -1.0
h = matrix(0.0, (n,1))
A = matrix(1.0, (1,n))
b = matrix(1.0)

# Compute trade-off.
N = 100
mus = [ 10**(5.0*t/N-1.0) for t in range(N) ]
portfolios = [ qp(mu*S, -pbar, G, h, A, b)['x'] for mu in mus ];
returns = [ dot(pbar,x) for x in portfolios ]
risks = [ sqrt(dot(x, S*x)) for x in portfolios ]
```

```python
figure()
ax = gca()
ax.plot(risks, returns, '*')
ax.set_xlabel('standard deviation')
ax.set_ylabel('expected return')
#ax.set_axis([0, 0.2, 0, 0.15])
#ax.set_title('Risk-return trade-off curve (fig 4.12)')
#ax.set_yticks([0.00, 0.05, 0.10, 0.15])
```

```python
figure()
xax = returns
ax = gca()
c1 = [ x[0] for x in portfolios ]
c2 = [ x[0] + x[1] for x in portfolios ]
c3 = [ x[0] + x[1] + x[2] for x in portfolios ]
c4 = [ x[0] + x[1] + x[2] + x[3] for x in portfolios ]
ax.fill_between(xax, 0, c1, facecolor='#F0F0F0')
ax.fill_between(xax, c1, c2,  facecolor = '#D0D0D0')
ax.fill_between(xax, c2, c3,  facecolor = '#F0F0F0')
ax.fill_between(xax, c3, c4,  facecolor = '#D0D0D0')
ax.set_xlabel('standard deviation')
ax.set_ylabel('allocation')
#pylab.text(.15,.5,'x1')
#pylab.text(.10,.7,'x2')
#pylab.text(.05,.7,'x3')
#pylab.text(.01,.7,'x4')
ax.set_title('Optimal allocations (fig 4.12)')
```
