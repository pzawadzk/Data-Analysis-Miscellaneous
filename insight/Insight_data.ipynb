```python
X = df.values[:,:-1]
y = df.values[:,-1]
```

```python
from sklearn import neighbors
from sklearn import preprocessing
from sklearn.decomposition import PCA, KernelPCA
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import *
from sklearn.svm import SVR, NuSVR
from sklearn import cross_validation
```

```python
np.shape(X)
```

```python
def sklearn_models(method='forest', **kwargs):
    if   method == "tree":       model = DecisionTreeRegressor()
    elif method == "forest":     model = RandomForestRegressor(**kwargs)
    elif method == "bayesridge": model = BayesianRidge()
    elif method == "ard":        model = ARDRegression()
    elif method == "lars":       model = Lars()
    elif method == "lasso":      model = Lasso()
    elif method == "linear":     model = LinearRegression()
    elif method == "passiveagressive": model = PassiveAggressiveRegressor()
    elif method == "sgd":        model = SGDRegressor()
    elif method == "svr":        model = SVR(**kwargs)
    elif method == "nusvr":      model = NuSVR(**kwargs)
    elif method == "knn":        model = neighbors.KNeighborsRegressor(5, weights="distance", metric="minkowski")

    else: print "method not implemented"; XX
    return model
```

### Cross validation, Feature normalization

```python
def get_cv_error(kfold, X, y, model, returnlist=False):
    MAE = []
    for train, test in kfold:
        X_train = X[train]
        y_train = y[train]
        X_test = X[test]
        y_test = y[test]

        scaler = preprocessing.StandardScaler().fit(X_train)
        X_train = scaler.transform(X_train)
        X_test = scaler.transform(X_test)

        model.fit(X_train, y_train)
        error = np.sum(np.abs(model.predict(X_test) - y_test)) / len(y_test)
        MAE.append(error)

    if returnlist:
        return MAE
    else:
        return np.sum(MAE) / len(kfold)
```

### Cross validation, Stratified sampling

```python
def get_learning_errors(method, X, y, train_sizes):

    MAE = []
    sizes = []
    for train_size in train_sizes:
        X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=1-train_size, random_state=0)
        kfold = cross_validation.StratifiedKFold(y_train, n_folds=5, shuffle=True)
        MAE.append(get_cv_error(kfold, X_train, y_train, sklearn_models(method), returnlist=True))
        sizes.append(X_train.shape[0])
    return sizes, MAE
```

###Ploting learning curve

```python
def plot_learning_curve(plt, method, X, y, label, ll, train_sizes=np.linspace(.1, 1.0, 5)):

    train_sizes, test_scores = get_learning_errors(method, X, y, train_sizes=train_sizes)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color=ll[2])
    plt.plot(train_sizes, test_scores_mean, ll[:2], color=ll[2],
             label=label)

    return  plt
```

```python
# knn, forest, svr(nusvr) are the best
ll = ['o-k', 's-r', '+-b', 'o-m', 's-y', '+-g', '^-r']

figure(figsize=(8,6))
for  ii, method in enumerate([#"krr", 
       # "linear", #"passiveagressive", "sgd", 
        "tree", 
        "forest", #"bayesridge", #"ard", "lars", "lasso", 
        "svr", 
        #"nusvr", 
        #"knn", 
        ]):

    plt = plot_learning_curve(plt, method, X, y, method, ll[ii])

    plt.grid()
    plt.xlabel("Training dataset", fontsize=16)
    plt.ylabel("MAE (eV)", fontsize=16)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
from atomic_constants import *
from collections import defaultdict, Counter
def get_X(names, elmap):
    Eatomicnum = []
    rad = []
    elecneg = []
    rowlist = []
    collist = []
    Eionization = []

    for name in names:
        Eatomicnum.append(atomic_number[name])
        rad.append(radius[name])
        elecneg.append(pauling[name])
        rowlist.append(row[name])
        collist.append(col[name])
        Eionization.append(Eion[name])

    val = [np.mean(Eatomicnum),
           np.mean(rad), np.max(rad)-np.min(rad), np.std(rad),
           np.mean(elecneg), np.max(elecneg)-np.min(elecneg), np.std(elecneg),
           np.mean(rowlist), np.std(rowlist),
           np.mean(collist), np.max(collist)-np.min(collist), np.std(collist),
           np.mean(Eionization), np.max(Eionization)-np.min(Eionization), np.std(Eionization)]

    elval = [0,] * len(elmap)
    uniqueel = Counter(names) # get unqiue elements
    for k1, v1 in uniqueel.items(): # propotion of each elements in cell
        elval[elmap[k1]] = float(v1) / len(names) # pauling[k1] / nn gives exactly the same name

    return val + elval
```
