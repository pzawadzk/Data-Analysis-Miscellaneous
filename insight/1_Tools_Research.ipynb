```python
import matplotlib.pyplot as plt
import numpy as np

def plot_phase_diagram(data_to_plot, species, range=[-1,-1], axis=[1,0], legend = False, showx=True, showy=True, name = False):
    fig = plt.figure(figsize=(4,6))
    ax = fig.gca()


    loc=list(np.linspace(-2,0,5))
    ax.xaxis.set_major_locator(plt.FixedLocator(loc))
    ax.yaxis.set_major_locator(plt.FixedLocator(loc))
    
    loc=list(np.linspace(-2,0,5))
    ax.xaxis.set_major_locator(plt.FixedLocator(loc))
    col = {'Cu2S3Sn1': 'BlanchedAlmond',  'Cu3S4Sn1':'Cyan', 'Cu4S16Sn7':'r', "Cu4S4Sn1": 'g', "Cu4S32Sn15":'m',  
           'Cu1S2Sb1': 'BlanchedAlmond',  'Cu3S4Sb1':'Cyan', 'Cu12S13Sb4':'r', "Cu3S3Sb1": 'g',  
            "Cu2S1":'DeepSkyBlue','Cu1S1':'y',
           "S1Sn1":'LightSeaGreen',  "S2Sn1":'DarkSalmon',
            "S3Sb2":'LightSeaGreen', 
            'S': 'w', 'Cu': 'w', 'Sn': 'w', 'Sb': 'w',
           "Cu1S1Sn1": 'k', "Cu1S2Sn1": 'm'
            }
    col = ['k', 'r', 'b', 'g', 'y', 'c', 'm', 'DarkSalmon', 'LightSeaGreen', 'BlanchedAlmond', 'DeepSkyBlue']*40
    for i, key in enumerate(data_to_plot.keys()):
        
        if len(data_to_plot[key][1])>1:
            ax.fill(data_to_plot[key][axis[0]], data_to_plot[key][axis[1]], col[i], alpha=0.75, label = key)
            
            #print i, key, col[i]
    if legend:
        ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        
    if showy:
      ax.set_ylabel(r'$\Delta\mu_{\mathrm{%s}}$ [eV]'%species[axis[1]], fontsize=25)
    else:    
        plt.setp(ax.get_yticklabels(), visible=False)
        
    if showx:
        ax.set_xlabel(r'$\Delta\mu_{\mathrm{%s}}$ [eV]'%species[axis[0]], fontsize=25)
    else:    
        plt.setp(ax.get_xticklabels(), visible=False)
        
    ax.axis(xmin=range[0], xmax=0, ymin =range[1], ymax=0)
    
    if name:
        savefig(name)
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
          fancybox=True, shadow=True, ncol=5)
```

```python
from sklearn import neighbors
from sklearn import preprocessing
from sklearn.decomposition import PCA, KernelPCA
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import *
from sklearn.svm import SVR, NuSVR
from sklearn import cross_validation
```

```python
def sklearn_models(method='forest', **kwargs):
    if   method == "tree":       model = DecisionTreeRegressor()
    elif method == "forest":     model = RandomForestRegressor(**kwargs)
    elif method == "bayesridge": model = BayesianRidge()
    elif method == "ard":        model = ARDRegression()
    elif method == "lars":       model = Lars()
    elif method == "lasso":      model = Lasso()
    elif method == "linear":     model = LinearRegression()
    elif method == "passiveagressive": model = PassiveAggressiveRegressor()
    elif method == "sgd":        model = SGDRegressor()
    elif method == "svr":        model = SVR(**kwargs)
    elif method == "nusvr":      model = NuSVR(**kwargs)
    elif method == "knn":        model = neighbors.KNeighborsRegressor(5, weights="distance", metric="minkowski")

    else: print "method not implemented"; XX
    return model
```

### Cross validation, Stratified sampling

```python
def get_cv_error(kfold, X, y, model, returnlist=False):
    MAE = []
    for train, test in kfold:
        X_train = X[train]
        y_train = y[train]
        X_test = X[test]
        y_test = y[test]

        scaler = preprocessing.StandardScaler().fit(X_train)
        X_train = scaler.transform(X_train)
        X_test = scaler.transform(X_test)

        model.fit(X_train, y_train)
        error = np.sum(np.abs(model.predict(X_test) - y_test)) / len(y_test)
        MAE.append(error)

    if returnlist:
        return MAE
    else:
        return np.sum(MAE) / len(kfold)
    
def get_learning_errors(method, X, y, train_sizes):

    MAE = []
    sizes = []
    for train_size in train_sizes:
        X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=1-train_size, random_state=0)
        kfold = cross_validation.StratifiedKFold(y_train, n_folds=5, shuffle=True)
        MAE.append(get_cv_error(kfold, X_train, y_train, sklearn_models(method), returnlist=True))
        sizes.append(X_train.shape[0])
    return sizes, MAE

df['energy'].hist()
ax = gca()
ax.set_xlabel('energy')
ax.set_ylabel('Number of data points')
```

### Learning Curve

```python
def plot_learning_curve(plt, method, X, y, label, ll, train_sizes=np.linspace(.1, 1.0, 5)):

    train_sizes, test_scores = get_learning_errors(method, X, y, train_sizes=train_sizes)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color=ll[2])
    plt.plot(train_sizes, test_scores_mean, ll[:2], color=ll[2],
             label=label)

    return  plt

# knn, forest, svr(nusvr) are the best
ll = ['o-k', 's-r', '+-b', 'o-m', 's-y', '+-g', '^-r']

figure(figsize=(8,6))
for  ii, method in enumerate([#"krr", 
       # "linear", #"passiveagressive", "sgd", 
        "tree", 
        "forest", #"bayesridge", #"ard", "lars", "lasso", 
        "svr", 
        #"nusvr", 
        #"knn", 
        ]):

    plt = plot_learning_curve(plt, method, X, y, method, ll[ii])

    plt.grid()
    plt.xlabel("Training dataset", fontsize=16)
    plt.ylabel("MAE (eV)", fontsize=16)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
```
