```python
drivers = np.loadtxt('drivers.txt', dtype=np.int16)
ras=[]
for dr in drivers[2:50]:
    print '#######', dr
    val = []
    figure()
    for i in range(1,21):
        d = np.genfromtxt('drivers/%d/%d.csv'%(dr,i), delimiter=',', skip_header=True)
        dd = get_distance(d)
        v = np.gradient(dd)
        nstamp = 10
        dist = [100, 500,] + [1000*i for i in range(1,nstamp-1)]
        time = np.zeros(nstamp)-1
        velo = np.zeros(nstamp)-1
        j = 0
        for i, d in enumerate(dd):
            #print d, dist[j]
            if d > dist[j]:
                time[j] = i
                velo[j] = v[i]
                j += 1
            if j>=len(dist):
                break
        print time
```
